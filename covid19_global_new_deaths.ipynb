```python
import pandas as pd
```

```python
def global_new_deaths():
    df = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
    df = df.drop(['Country/Region', 'Province/State', 'Lat', 'Long'], axis=1).sum(axis=0).diff().fillna(value=0).rolling(7, min_periods=1).mean()
    return df
```

```python
def new_deaths_by_country(country=None):
    df = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
    if not country is None:
        df = df.loc[df['Country/Region'] == country]
    df = df.drop(['Country/Region', 'Province/State', 'Lat', 'Long'], axis=1).sum(axis=0).diff().fillna(value=0).rolling(7, min_periods=1).mean()
    return df
```

```python
df = global_new_deaths()
df.plot(title='Global Deaths')
```
