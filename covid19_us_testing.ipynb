```python
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```python
def datetime_to_str(dtime):
    return '{0}-{1}-{2}'.format(str(dtime.month).zfill(2), str(dtime.day).zfill(2), str(dtime.year).zfill(4))
# datetime_to_str(datetime.now())
```

```python
def get_tests_data(dtime):
    exclude = [
        'Diamond Princess',
        'Grand Princess',
        'American Samoa',
        'Virgin Islands',
        'Northern Mariana Islands'
    ]
    file = '{0}.csv'.format(datetime_to_str(dtime))
    df = pd.read_csv('csse_covid_19_data/csse_covid_19_daily_reports_us/{0}'.format(file),
                     usecols=['Province_State', 'Total_Test_Results']).fillna(0)
    df = df[~df.Province_State.isin(exclude)] # not isin
    df.rename(columns={ 'Total_Test_Results': datetime_to_str(dtime) }, inplace=True)
    df.set_index(['Province_State'], inplace=True)
    return df

def get_daily_tests_timeseries(dtime, ndays=30):
    df = get_tests_data(dtime)
    for days in range(1, ndays+1):
        df_2 = get_tests_data(dtime - timedelta(days=days))
        df = pd.concat([df_2, df], axis=1, join='inner')
    return df

def get_daily_new_tests_timeseries(dtime, ndays=30):
    df = get_daily_tests_timeseries(dtime, ndays=ndays)
    return df.diff(axis=1).fillna(0).sum(axis=0)

def get_daily_new_tests_by_state_timeseries(dtime, ndays=30):
    df = get_daily_tests_timeseries(dtime, ndays=ndays)
    return df.diff(axis=1).fillna(0)
```

```python
def get_hospitalized_data(dtime):
    exclude = [
        'Diamond Princess',
        'Grand Princess',
        'American Samoa',
        'Virgin Islands',
        'Northern Mariana Islands'
    ]
    file = '{0}.csv'.format(datetime_to_str(dtime))
    df = pd.read_csv('csse_covid_19_data/csse_covid_19_daily_reports_us/{0}'.format(file),
                     usecols=['Province_State', 'People_Hospitalized']).fillna(0)
    df = df[~df.Province_State.isin(exclude)] # not isin
    df.rename(columns={ 'People_Hospitalized': datetime_to_str(dtime) }, inplace=True)
    df.set_index(['Province_State'], inplace=True)
    return df

def get_daily_hospitalized_timeseries(dtime, ndays=30):
    df = get_hospitalized_data(dtime)
    for days in range(1, ndays+1):
        df_2 = get_hospitalized_data(dtime - timedelta(days=days))
        df = pd.concat([df_2, df], axis=1, join='inner')
    return df

def get_daily_new_hospitalized_timeseries(dtime, ndays=30):
    df = get_daily_hospitalized_timeseries(dtime, ndays=ndays)
    return df.diff(axis=1).fillna(0).sum(axis=0)

def get_daily_new_hospitalized_by_state_timeseries(dtime, ndays=30):
    df = get_daily_hospitalized_timeseries(dtime, ndays=ndays)
    return df.diff(axis=1).fillna(0)
```

```python
def get_daily_data(dtime, col):
    exclude = [
        'Diamond Princess',
        'Grand Princess',
        'American Samoa',
        'Virgin Islands',
        'Northern Mariana Islands',
        'Puerto Rico',
    ]
    file = '{0}.csv'.format(datetime_to_str(dtime))
    df = pd.read_csv('csse_covid_19_data/csse_covid_19_daily_reports_us/{0}'.format(file),
                     usecols=['Province_State', col]).fillna(0)
    df = df[~df.Province_State.isin(exclude)] # not isin
    df.set_index(['Province_State'], inplace=True)
    return df

def get_daily_confirmed(dtime):
    conf_curr = get_daily_data(dtime, 'Confirmed')
    conf_prev = get_daily_data(dtime - timedelta(days=1), 'Confirmed')
    conf = (conf_curr - conf_prev)
    conf.rename(columns={'Confirmed': datetime_to_str(dtime)}, inplace=True)
    return conf

def get_daily_tested(dtime, lag=0):
    tstd_curr = get_daily_data(dtime - timedelta(days=lag), 'Total_Test_Results')
    tstd_prev = get_daily_data(dtime - timedelta(days=lag+1), 'Total_Test_Results')
    tstd = (tstd_curr - tstd_prev)
    tstd.rename(columns={'Total_Test_Results': datetime_to_str(dtime)}, inplace=True)
    return tstd

def get_confirmed_timeseries(dtime, ndays=10):
    df = get_daily_confirmed(dtime)
    for days in range(1, ndays+1):
        df_2 = get_daily_confirmed(dtime - timedelta(days=days))
        df = pd.concat([df_2, df], axis=1, join='inner')
    return df.rolling(7, axis=1, min_periods=1).mean()

def get_tested_timeseries(dtime, lag=0, ndays=10):
    df = get_daily_tested(dtime, lag=lag)
    for days in range(1, ndays+1):
        df_2 = get_daily_tested(dtime - timedelta(days=days), lag=lag)
        df = pd.concat([df_2, df], axis=1, join='inner')
    return df.rolling(7, axis=1, min_periods=1).mean()
    
def get_pct_plus(dtime, ndays=10, lag=0):
    conf = get_confirmed_timeseries(dtime, ndays=ndays)
    tstd = get_tested_timeseries(dtime, ndays=ndays)
    return 100 * conf / tstd

# get_daily_confirmed(datetime.now() - timedelta(days=2))
# get_daily_tested(datetime.now() - timedelta(days=2))
# get_confirmed_timeseries(datetime.now() - timedelta(days=2))
# get_tested_timeseries(datetime.now() - timedelta(days=2))
# get_pct_plus(datetime.now() - timedelta(days=2), lag=1)
# get_pct_plus(datetime.now() - timedelta(days=2), lag=2)
# get_pct_plus(datetime.now() - timedelta(days=2), lag=3)
# get_pct_plus(datetime.now() - timedelta(days=2), lag=4)
get_pct_plus(datetime.now() - timedelta(days=1), ndays=10, lag=0)
```
