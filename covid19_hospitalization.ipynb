```python
"""
  Datasource: https://covidtracking.com/data
"""
import pandas as pd
```

```python
def daily_us_hospitalization(state=None):
    df = pd.read_csv('hospitalization/all-states-history.csv',
                    usecols=['date', 'state', 'hospitalizedCurrently']).fillna(0)
    if not state is None:
        df = df.loc[df['state'] == state]
    df.sort_values('date', ascending=True, inplace=True)
    df = df.groupby(['date']).sum()
    return df
```

```python
df = daily_us_hospitalization()
df.plot(title='USA')
```
